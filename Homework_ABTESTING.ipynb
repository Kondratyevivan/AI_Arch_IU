Дана таблица с данными о температуре (град С) и давлении (МПа) внутри ректификационной колонны, в которой находится жидкость и газ. Температура регулируется подачей холодной воды в кубовую часть(низ колонны), а давление регулируется открытием клапана сдувки лишнего газа сверху колонны. По регламенту, давление в колонне не должно быть ниже 2 Мпа.

```python

print ('Проверка гипотезы о том, что после настройки регуляторов количество нарушений режима по давлению свелось к нулю:')
hyp = True
firstIndex = 0
#Определяем индекс строки, в которой была произведена настройка регуляторов
for j in range(len(df.index)):
  if df.iloc [j]['Unnamed: 3'] == 'Настройка регуляторов':
    firstIndex = j
#Проверяем каждое значение после настройки, ниже ли оно минимального порога
for i in range(firstIndex,len(df.index)):
  if df.iloc [i]['Press']<2:
    hyp = False
#Если ни одно значение не выходит за рамки, значит гипотеза верна
if hyp == True:
  print ('Гипотеза верна')
```

```python
print('Проверка гипотезы о том, что колебательность температуры после настройки снизилась более, чем на 20%')
#Разделяем таблицу на 2, до и после настройки регуляторов
df1 = df.iloc [:firstIndex]
df2 = df.iloc [firstIndex:]
#Определяем стандатрное отклонение температуры до настройки
devBefore = df1['Temp'].std()
devAfter = df2['Temp'].std()
print('Отклонение температуры до настройки: ',devBefore)
print('Отклонение температуры после настройки настройки: ',devAfter)
devRatio =100 - devAfter*100/devBefore
print('Отклонение после настройки на ', devRatio,'% ниже отклонения до настройки')
if devAfter*100/devBefore>20:
  print('Гипотеза верна')
```
